```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
from datetime import datetime
```

```python
r=0
while 1:
    #creat 500*500 image and and circle in half
    img = np.zeros((500,500,3),np.uint8)
    cv2.circle(img,(250,250),100,(0,0,255),1)
    
     #To print Date:
    now = datetime.now()   
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img, f'Date : {now.year}-{now.month}-{now.day}', (0,30), font, 0.70, (245, 244, 18), 2, cv2.LINE_AA)
    #cv2.imshow('Circle', img)
#===============================================================================================================    
    #TO print time Now:
    now = datetime.now()
    s = datetime.now().strftime("%S")  #(%s) To get The second from the date 
    m = datetime.now().strftime("%M")  #(%M) To get The Minutes from the date
    h = datetime.now().strftime("%I")  #(%I) To get The Hour from the date
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img, f'Time : {h}:{m}:{s}', (0,65), font, 0.70, (245, 244, 18), 2, cv2.LINE_AA)
    #cv2.imshow('Circle', img)
#===============================================================================================================
    #for loop for hours circls
    x = int(250+100*math.cos(math.radians(r)))
    y = int(250+100*math.sin(math.radians(r)))

    for i in range(12):
        x = int(250+100*math.cos(math.radians(r)))
        y = int(250+100*math.sin(math.radians(r)))
        cv2.circle(img,(x,y),4,(0,255,255),-1)
        r+=30  # 360 /12=30
#===============================================================================================================        
    #for loop for min cir
    for i in range(60):
        x = int(250+100*math.cos(math.radians(r)))
        y = int(250+100*math.sin(math.radians(r)))
        cv2.circle(img,(x,y),2,(0,255,255),-1)
        r+=6     # 360/60=6
#===============================================================================================================
    # This for (second hand):-
    now = datetime.now()
    s = datetime.now().strftime("%S")
    SecondAngle = ((int(s) * 6) - 90)
    x = int((250 + 90*math.cos(math.radians(SecondAngle))))
    y = int((250 + 90*math.sin(math.radians(SecondAngle))))
    cv2.line(img, (250, 250), (x, y), (255, 255, 0))
#=================================================================================================================
     # This for (Minutes hand):-
    now = datetime.now()
    minute = datetime.now().strftime("%M")
    MinutesAngle = (((int(minute) * 6) - 90))
    x = int((250 + 70*math.cos(math.radians(MinutesAngle))))
    y = int((250 + 70*math.sin(math.radians(MinutesAngle))))
    cv2.line(img, (250, 250), (x, y), (110, 200, 10), 2)
#=================================================================================================================    
    # This for (Hour hand):-
    now = datetime.now()
    hour = datetime.now().strftime("%I")
    HourAngle = (360//12*((now.hour+(now.minute/60))))+270
    x = int((250 + 50*math.cos(math.radians(HourAngle))))
    y = int((250 + 50*math.sin(math.radians(HourAngle))))
   
    cv2.line(img, (250, 250), (x, y), (5, 100, 100), 3)

    cv2.imshow('Circle',img)
    k = cv2.waitKey(1000)
    if k == ord('q'):
        break

cv2.destroyAllWindows()
```
